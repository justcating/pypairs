```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module2/data/sales_data.csv", index_col="Unnamed: 0")
```

# Завдання 4
Яка загальна виручка у продавця Luis?

```python
luis_sales = df[df['salesman'] == 'Luis']['sale_amt'].sum()
print("Загальна виручка Luis:", luis_sales)
```

# Завдання 5

Скільки і якого товару було продано?

див groupby(),

```python
grouped_sales = df.groupby('item')['units'].sum()
print(grouped_sales)
```

# Завдання 6
Перевірте чи правильно розраховано загальну вартість покупки. Виведіть ті рядки де є помилка. Виправте дані в таблиці

```python
incorrect_sales = df[df['sale_amt'] != df['units'] * df['unit_price']]
print("Неправильні розрахунки:")
print(incorrect_sales)
```

```python
df.loc[df['sale_amt'] != df['units'] * df['unit_price'], 'sale_amt'] = df['units'] * df['unit_price']
print("Виправлені дані:")
df
```
